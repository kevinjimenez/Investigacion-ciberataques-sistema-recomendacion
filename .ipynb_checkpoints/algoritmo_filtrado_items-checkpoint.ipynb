```python
import os
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
path_datos_raiting = '/home/dev-11/Documents/kevin/Investigacion-ciberataques-sistema-recomendacion/datos-prueba/raiting_25.csv'
path_datos_ataques = '/home/dev-11/Documents/kevin/Investigacion-ciberataques-sistema-recomendacion/datos-prueba/datos_valores_25.csv'
```

```python
# leer archivo de ratings por usuario
df = pd.read_csv(path_datos_raiting, sep=',', names=['user_id','item_id','rating'])
```

```python
# leer archivo de titulos de peliculas con su id
titulos_peliculas = pd.read_csv(path_datos_nombre_categoria)
```

```python
# listar 10 primeros registros del archivo de titulos de peliculas
titulos_peliculas.head(10)
```

```python
# union de los dataframes de ratings con los titulos de peliculas con su id  
df = pd.merge(df, titulos_peliculas, on='item_id')
```

```python
df.head()
```

```python
# los raitings agrupados por titulo y calculando 
# la media x su raiting media o promedio de cada pelicula
ratings = pd.DataFrame(df.groupby('name_x')['rating'].mean())
```

```python
ratings.head()
```

```python
ratings['numero_de_rating'] = df.groupby('name_x')['rating'].count()
ratings.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
ratings['rating'].hist(bins=50)
```

```python
ratings['numero_de_rating'].hist(bins=60)
```

```python
import seaborn as sns
sns.jointplot(x='rating', y='numero_de_rating', data=ratings)
```

```python
matriz_pelicula = df.pivot_table(index='user_id', columns='name_x', values='rating')
matriz_pelicula.head(10)
```

```python
ratings.sort_values('numero_de_rating', ascending=False).head(5)
```

```python
user_rating = matriz_pelicula['CWE-434: Unrestricted Upload of File with Dangerous Type']
```

```python
user_rating.head()
```

```python
similar=matriz_pelicula.corrwith(user_rating)
```

```python
similar.head()
```

```python
corr_AFO = pd.DataFrame(similar, columns=['correlation'])
corr_AFO.dropna(inplace=True)
corr_AFO.head()
```

```python
corr_AFO = corr_AFO.join(ratings['numero_de_rating'],how='left', lsuffix='_left', rsuffix='_right')
corr_AFO.head(5)
```

```python
corr_AFO[corr_AFO['numero_de_rating'] > 2].sort_values(by='correlation', ascending=False).head(10)
```
